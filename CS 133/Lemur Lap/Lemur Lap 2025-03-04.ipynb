```python
import pandas as pd
l = [ x for x in (2,4,6,8,10) for _ in range(10)]
```

```python
l = [ x for x in (2,4,6,8,10) * 10]
```

```python
print(l)
```

```python
s = pd.DataFrame(l).sample(frac=1).reset_index(drop = True)
```

```python
s.describe()
```

```python
s.hist(bins = 19)
```

```python
sample = [s.sample(n=30).mean() for i in range (10000)]
```

```python
pd.DataFrame(sample).plot(kind = 'hist', bins = 19, title= f"n=1")

#wrap above in a function that takes in n and that takes in how many experiments (1000) 
```
