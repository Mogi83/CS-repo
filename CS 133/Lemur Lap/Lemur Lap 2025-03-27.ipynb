```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")

penguins.info()
```

```python
sns.barplot(data=penguins, x="island", y="body_mass_g", hue="sex")
```

```python
sns.histplot(data=penguins, x="island", hue="sex", weights="body_mass_g", multiple="stack")
```

```python
dots = sns.load_dataset("dots")

dots.info()
```

```python
dots.groupby('choice').count()
```

```python
sns.swarmplot(dots, x='choice',y='firing_rate')
```
