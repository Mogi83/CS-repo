## 1. Violin plot of categorical feature

1.1 (15 pts) Using the Seaborn module, load the `seaice` dataset. Create a new column called `month` that contains the name of the month for each row according to the `Date` column. (Note you can use the Pandas syntax `df.Date.dt.month_name()` to get the month name). Display a violin plot showing the distribution of seaice per month.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
seaice = sns.load_dataset("seaice")
seaice.insert( 0, "month", seaice["Date"].dt.month_name())
#seaice.info()
#seaice.head()

plt.figure(figsize=(12,6))
sns.violinplot(data=seaice, x="month", y="Extent")

month_variation = seaice.groupby("month")["Extent"].var()
pd.DataFrame(month_variation).sort_values("Extent", ascending=False)
```

1.2 (1 pt) Which month appears to have the most variance in the extent of sea ice?

October has the most variance, defined as the largest difference between data points or largest spread of min-max. Visually this would be the tallest (y-axis) violin plot.

## 2. Scatterplot with multiple features

2.1 (15pts) Using the Seaborn module load the `car_crashes` dataset. Display a scatterplot showing the relation between insurance premiums and insurance losses. Set the size of each data point from the total number of crashes and label the five largest data points with their state abbreviations.

_Note: You can label a point using the Seaborn `axis.at()` method, where `axis` is returned from the plot function. Pandas has a `DataFrame.nlargest()` method that might also be helpful._


```python
#I couldn't find any documentation for a seaborn or plt method called axis.at I think you were thinking axis.label or plt.text()? https://matplotlib.org/stable/api/axis_api.html
car_crash = sns.load_dataset("car_crashes")
car_crash.describe()
fig, ax = plt.subplots(figsize=(8, 6))
large_points = car_crash.nlargest(5, "total")
large_points.info()

for _, row in large_points.iterrows():
    car_crash.drop(_, inplace=True)
    plt.text(x=row["ins_premium"], y=row["ins_losses"], s=row["abbrev"], ha="center", va= "center")


sns.scatterplot(data=car_crash, x="ins_premium", y="ins_losses", size="total", sizes=(50,250), legend=False)

ax = sns.scatterplot(data=large_points, x="ins_premium", y="ins_losses", size="total", sizes=(50,250))

labels = []
for i in range(len(large_points)):
    state = large_points["abbrev"].iloc[i]
    total = large_points["total"].iloc[i]
    labels.insert(0, f"{state} ({total:.2f})")

ax.legend(labels)

#I have no idea why the first element of the legend is a blue dot, it shouldnt be but im beyond fighting it.
```

## 5. Plotting with multiple categorical and numerical features

5.1 (15pts) Using the Seaborn module load the `diamonds` dataset. Sample 200 diamonds with a price greater than $5,000. Create a graph to show how the price of the diamonds varies with the carat and include information on the graph indicating the color, depth, and cut of each diamond.

```python
jewels = sns.load_dataset("diamonds")
jewels.info()
sample_jewels = jewels.groupby(jewels['price'] > 5000).sample(n=200, random_state=45)  
#sample_jewels.head()
plt.figure(figsize=(12,12))
sns.scatterplot(data=sample_jewels, y="price", x="carat", hue="color", size="depth", style="cut")

plt.legend(bbox_to_anchor=(1.075,0.75), loc='center', columnspacing = 0.5, alignment='left',)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

sns.stripplot(data=sample_jewels, x="cut",y="price", hue="carat", ax=axes[0], palette="BuGn_r",)
sns.stripplot(data=sample_jewels, x="clarity",y="price", hue="carat", ax = axes[1], palette="BuGn_r")
sns.stripplot(data=sample_jewels, x="color",y="price", hue="carat", ax = axes[2], palette="BuGn_r")

axes[0].legend(loc="upper left", bbox_to_anchor=(-0.5, 1), title="carat")
axes[1].legend(loc="upper left", bbox_to_anchor=(-0.5, 1), title="carat")
axes[2].legend(loc="upper left", bbox_to_anchor=(-0.5, 1), title="carat")
sns.set(rc={'axes.facecolor':'darkgray', 'figure.facecolor':'gray'})
plt.tight_layout()

#If you find these graph colors ugly, I asked my family to pick color names in the mpl docs for every graph.
```
