#### *Which countries are most prone to UFO sightings?*

List the counts of UFO observations by country

```python
population_data = {
    'us': 318.4,
    'ca': 35.43,
    'gb': 64.6,
    'au': 23.48,
    'de': 80.98
}

ufo_2014 = ufo[ufo["date"].dt.year == 2014]
ufo_count_by_country = ufo_2014.groupby("country").size().reset_index(name="ufo_count")

population_df = pd.DataFrame(list(population_data.items()), columns=["country", "population"])
ufo_with_population = pd.merge(ufo_count_by_country, population_df, on="country", how="inner")

ufo_with_population["ufo_per_capita"] = ufo_with_population["ufo_count"] / ufo_with_population["population"]
```

Since countries have different populations it is more informative to normalize the above data _per capita_. Create a chart showing the number of observations by country per capita for the the year 2014. Make sure to label the axes.

The population for each country in 2014 is given below (in millions):

| Country | Population in 2014 (millions) |
|:-------|:-----:|
| US| 318.4 |
| Canada| 35.43 |
| Great Britain| 64.6 |
| Australia| 23.48 |
| Germany (de)| 80.98 |

```python
plt.figure(figsize=(10, 6))
plt.bar(ufo_with_population["country"], ufo_with_population["ufo_per_capita"])

plt.xlabel('Country')
plt.ylabel('UFO Sightings per Capita (2014)')
plt.title('UFO Sightings per Capita by Country in 2014')
plt.xticks(rotation=45)  

plt.tight_layout()
plt.show()
```

#### _How has the number of observations changed over time?_

Create a visualization of the number of reported UFO sightings worldwide by year and country, beginning in 1940. Make sure the graph is visually pleasing and has proper labels on the axes and a legend.

```python
ufo = ufo[ufo["date"].dt.year >= 1940]
ufo_count_by_year_country = ufo.groupby([ufo["date"].dt.year, "country"]).size().reset_index(name="sightings")
ufo_count_by_year_country.columns = ['year', 'country', 'sightings']
ufo_pivot = ufo_count_by_year_country.pivot_table(index='year', columns='country', values='sightings', aggfunc='sum', fill_value=0)


plt.figure(figsize=(12, 8))
ufo_pivot.plot(kind='line', figsize=(12, 8))

plt.xlabel('Year')
plt.ylabel('Number of UFO Sightings')
plt.title('Reported UFO Sightings Worldwide by Year and Country (From 1940)')
plt.xticks(rotation=45)  
plt.legend(title='Country')
plt.tight_layout()

plt.axvline(x=1990, color='red', linestyle='--', label='X-Files Debut (1990)')
plt.annotate('X-Files Debut', xy=(1990, 10), xytext=(1992, -200), arrowprops=dict(facecolor='red', shrink=0.05), fontsize=12, color='red')

plt.show()
```

For the next series of visualizations we will want to look at observations by month, day, and hour. To simplify ordering the graphs correctly we will use the `calendar` library in Python.

Import the `calendar` library and create the variables `month_list` and `weekday_list` from the `calendar` properties `month_name` and `day_name`. _Note: The `month_name` list starts with an empty string so you will need to slice it starting at index 1_.

```python
import calendar as cl

month_list = list(cl.month_name[1:])
weekday_list =list(cl.day_name[::])
```

Using the `date` feature of the dateframe, create a new column called `hour` from the `datetime.dt.hour` property.

```python
ufo['hour'] = ufo["date"].dt.hour
ufo.head()
```

Create a new column called `month` which contains the month name from the `date` feature. Use the `datetime.dt.month_name()` method.

```python
ufo['month'] = ufo["date"].dt.month_name()
ufo.head()
```

---

#### _Which hour of the day are UFO sightings most commonly reported?_

Create a visualization that graphs the number of UFO observations per hour for each month of the year. The graph legend should be the names of the months in calendar order and the axes should be labeled appropriately.

```python
counts = ufo.groupby(["month", "hour"]).date.count()
for month in month_list:
    ax = sns.lineplot(counts[month])
    
plt.legend(month_list)
plt.xlabel('Hour of Observation')
plt.ylabel('Count')
plt.title('Hour reported')
```

---

Create a visualization that graphs the number of UFO observations per weekday for each month of the year. The X axis should be the days of the week in calendar order. The graph legend should be the names of the months in calendar order and the axes should be labeled appropriately.
